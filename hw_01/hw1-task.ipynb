***

```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

df['LIMIT_BAL_CAT'] = df.apply(lambda row: str(np.array(['A','B','C','D','E','F'])[[row['LIMIT_BAL'] <= 1e4,
                                            (row['LIMIT_BAL'] > 1e4) & (row['LIMIT_BAL'] <= 1e5),
                                            (row['LIMIT_BAL'] > 1e5) & (row['LIMIT_BAL'] <= 2e5), 
                                            (row['LIMIT_BAL'] > 2e5) & (row['LIMIT_BAL'] <= 4e5),
                                            (row['LIMIT_BAL'] > 4e5) & (row['LIMIT_BAL'] <= 7e5),
                                            row['LIMIT_BAL'] > 7e5]][0]), axis=1)
```

```python
# проверка
df.groupby('LIMIT_BAL_CAT')['LIMIT_BAL'].aggregate([np.min, np.max])
```

***

```python
# (8) построить зависимость кредитного лимита от образования для каждого пола
# график необходимо сделать очень широким (на весь экран)
with plt.xkcd():
    fig, ax = plt.subplots(figsize=(15,5))
    
    x1 = df[df['SEX']==1].groupby(['EDUCATION'])['LIMIT_BAL'].median().index
    y1 = df[df['SEX']==1].groupby(['EDUCATION'])['LIMIT_BAL'].median().values
    
    x2 = df[df['SEX']==2].groupby(['EDUCATION'])['LIMIT_BAL'].median().index
    y2 = df[df['SEX']==2].groupby(['EDUCATION'])['LIMIT_BAL'].median().values

    ax.bar(x1-0.2, y1, width=0.3, label='MEN')
    ax.bar(x2+0.2, y2, width=0.3, label='WOMEN')
    
    ax.set_title('Median LIMIT_BAL vs SEX/EDUCATION')
    ax.set_xlabel('EDUCATION')
    ax.set_ylabel('LIMIT_BAL')
    
    plt.legend()
```

```python
# (8) построить зависимость кредитного лимита от образования для каждого пола
# график необходимо сделать очень широким (на весь экран)

with plt.xkcd():
    fig, ax = plt.subplots(figsize=(15,5))

    sns.boxplot(x='EDUCATION', y='LIMIT_BAL', data=df[['SEX','EDUCATION','LIMIT_BAL']], hue='SEX', ax=ax)
    
    ax.set_title('LIMIT_BAL distribution vs SEX/EDUCATION')
    ax.set_xlabel('EDUCATION')
    ax.set_ylabel('LIMIT_BAL')
    ax.set_ylim([0,7e5])
```

Разделим датасет на числовые и категориальные признаки:

```python
# объединим переменные BILL_ и PAY_ в переменную NET_
for i in range(1,7):
    df['NET_AMT'+str(i)] = df['BILL_AMT'+str(i)] - df['PAY_AMT'+str(i)]
```

```python
df_num = df[['LIMIT_BAL', 'AGE', 'NET_AMT1', 'NET_AMT2', 'NET_AMT3', 'NET_AMT4', 'NET_AMT5', 'NET_AMT6', 
             'default payment next month']]
```

```python
df_cat = df[['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6',
             'default payment next month']]
```

```python
# числовые признаки
warnings.filterwarnings('ignore')
g = sns.pairplot(df_num, diag_kind='kde', hue='default payment next month')
g.fig.get_children()[-1].set_bbox_to_anchor((0.5, 0.025, 0, 0))
g.fig.subplots_adjust(bottom=0.08);
```

```python
# категориальные признаки (1)
g2 = sns.pairplot(df_cat, hue='default payment next month')
g2.fig.get_children()[-1].set_bbox_to_anchor((0.5, 0.025, 0, 0))
g2.fig.subplots_adjust(bottom=0.08);
```

```python
# категориальные признаки (2)
g3 = sns.PairGrid(df_cat, hue='default payment next month',
                 x_vars=["SEX", "EDUCATION", "MARRIAGE"],
                 y_vars=['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6'],
                 aspect=2, size=3)
g3.map(sns.barplot);
```
