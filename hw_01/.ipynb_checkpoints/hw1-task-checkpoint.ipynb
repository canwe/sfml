```python
# (3) посчитать число женщин с университетским образованием
# SEX (1 = male; 2 = female). 
# EDUCATION (1 = graduate school; 2 = university; 3 = high school; 4 = others). 

df[(df.SEX == 2)&(df.EDUCATION == 2)].shape[0]
```

```python
# (4) Сгрупировать по "default payment next month" и посчитать медиану для всех показателей начинающихся на BILL_ и PAY_
#TODO

df.groupby('default payment next month')[df.columns[df.columns.str.startswith(('PAY_','BILL_'))]].median().T
```

```python
# (5) постройте сводную таблицу (pivot table) по SEX, EDUCATION, MARRIAGE

df.pivot_table(values=df[['EDUCATION','MARRIAGE']], index=df['SEX'], aggfunc=pd.Series.mode)
```

```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

df['LIMIT_BAL_CAT'] = df.apply(lambda row: str(np.array(['A','B','C','D','E','F'])[[row['LIMIT_BAL'] <= 1e4,
                                            (row['LIMIT_BAL'] > 1e4) & (row['LIMIT_BAL'] <= 1e5),
                                            (row['LIMIT_BAL'] > 1e5) & (row['LIMIT_BAL'] <= 2e5), 
                                            (row['LIMIT_BAL'] > 2e5) & (row['LIMIT_BAL'] <= 4e5),
                                            (row['LIMIT_BAL'] > 4e5) & (row['LIMIT_BAL'] <= 7e5),
                                            row['LIMIT_BAL'] > 7e5]][0]), axis=1)
```

```python
# проверка
df.groupby('LIMIT_BAL_CAT')['LIMIT_BAL'].aggregate([np.min, np.max])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# (7) постироить распределение LIMIT_BAL (гистрограмму)

with plt.xkcd():
    fig, ax = plt.subplots(figsize=(16,5))
    ax.hist(df.LIMIT_BAL, bins=40)
    ax.set_title('LIMIT_BAL distribution')
```

```python
# (8) построить зависимость кредитного лимита от образования для каждого пола
# график необходимо сделать очень широким (на весь экран)
with plt.xkcd():
    fig, ax = plt.subplots(figsize=(15,5))
    
    x1 = df[df['SEX']==1].groupby(['EDUCATION'])['LIMIT_BAL'].median().index
    y1 = df[df['SEX']==1].groupby(['EDUCATION'])['LIMIT_BAL'].median().values
    
    x2 = df[df['SEX']==2].groupby(['EDUCATION'])['LIMIT_BAL'].median().index
    y2 = df[df['SEX']==2].groupby(['EDUCATION'])['LIMIT_BAL'].median().values

    ax.bar(x1-0.2, y1, width=0.3, label='MEN')
    ax.bar(x2+0.2, y2, width=0.3, label='WOMEN')
    
    ax.set_title('Median LIMIT_BAL vs SEX/EDUCATION')
    ax.set_xlabel('EDUCATION')
    ax.set_ylabel('LIMIT_BAL')
    
    plt.legend()
```

```python
# (8) построить зависимость кредитного лимита от образования для каждого пола
# график необходимо сделать очень широким (на весь экран)

with plt.xkcd():
    fig, ax = plt.subplots(figsize=(15,5))

    sns.boxplot(x='EDUCATION', y='LIMIT_BAL', data=df[['SEX','EDUCATION','LIMIT_BAL']], hue='SEX', ax=ax)
    
    ax.set_title('LIMIT_BAL distribution vs SEX/EDUCATION')
    ax.set_xlabel('EDUCATION')
    ax.set_ylabel('LIMIT_BAL')
    ax.set_ylim([0,7e5])
```

```python
# (9) построить зависимость кредитного лимита и образования только для одного из полов

with plt.xkcd(scale=2):
    fig, ax = plt.subplots(figsize=(16,5))
    sns.heatmap(data=pd.crosstab(df[df.SEX==1]['EDUCATION'],df[df.SEX==1]['LIMIT_BAL_CAT']), cmap='Blues', 
                linewidths=1, linecolor='grey', fmt="d", annot=True, vmax=5800)
    ax.set_title('LIMIT_BAL_CAT vs EDUCATION (SEX == 1)\n(counts)')
```
